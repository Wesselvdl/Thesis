```python
from openai import OpenAI
import pandas as pd

# Assuming df is your DataFrame and it has a column named 'data'
# Example DataFrame:
# df = pd.DataFrame({'data': ['First text string', 'Second text string', 'Third text string']})

# Initialize the OpenAI client with your API key
client = OpenAI(
    api_key="YOUR_API_KEY",
)

# Function to get embedding for a piece of text
def get_embedding(text):
    if text != " ":
        response = client.embeddings.create(
            input=text,
            model="text-embedding-ada-002"
        )
    return response.data[0].embedding

# Apply the function to each row in the 'data' column and store the result in a new column
df["Embedding"] = df["Description"].apply(get_embedding)
```
